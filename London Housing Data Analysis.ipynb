



##### 1.Convert the datatype of 'Date' column to Date_time Format/

```python
df.dtypes
```

```python
df['date']=pd.to_datetime(df['date'])
```

##### 2a.Add a new column 'year' in the df , which contains only years

```python
df['Year']=df['date'].dt.year
df
```

##### 2b.Add a new column "month" as 2nd column in the df, which contains only month.

```python
# df.insert(index, 'newcolumn_name',new_column_values)
df.insert(2,'Month',df.date.dt.month)
df
```

##### 3. Remove the column year and month from the df.

```python
df.drop(['Year','Month'], axis =1, inplace=True)
```
